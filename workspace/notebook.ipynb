```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
print(netflix_df.head())
print(netflix_df.columns)
```

```python
netflix_90 = netflix_df.query('release_year >= 1990 and release_year <= 1999')
#print(netflix_df['release_year'].head())

print(netflix_90.head())
```

```python
#find the most frequent movie duration
netflix_90['duration'].hist(bins=50)
```

```python
duration = int(netflix_90['duration'].median())
#duration2 =netflix_90['duration'].median()
print(duration)
```

```python
#subsetting by genre
print(netflix_df['genre'])

netflix_action = netflix_90[netflix_90['genre'] == 'Action']
print(netflix_action)
```

```python
#filter for action movies that last less than 90 minutes
less_than_90 = netflix_action.query('duration < 90')
print(less_than_90['duration'])
```

```python
#setting up a counter 
short_movie_count = 0

for lab, row in netflix_action.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

print(short_movie_count)
```
